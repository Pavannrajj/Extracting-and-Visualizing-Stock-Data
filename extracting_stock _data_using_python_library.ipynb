## Extracting Share Price

```python
apple_share_price = apple.history(period = "max")
```

```python
apple_share_price.head()
```

```python
apple_share_price.reset_index(inplace = True)
apple_share_price.head()
```

```python
apple_share_price.plot(x = 'Date',y = 'Open')
```
