We will extract Netflix stock data [https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html).


 Steps for extracting the data
1. Send an HTTP request to the web page using the requests library.
2. Parse the HTML content of the web page using BeautifulSoup.
3. Identify the HTML tags that contain the data you want to extract.
4. Use BeautifulSoup methods to extract the data from the HTML tags.
5. Print the extracted data


```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html"
```

```python
data = requests.get(url).text
print(data[:100])
```

```python
soup = BeautifulSoup(data,'html.parser')
```

```python
#creating empty data frame

netflix_data = pd.DataFrame(columns = ['Date',"Open","High","Low",'Close','Adj Close','Volume'])
```

```python
len(soup.find_all("tr"))
```

```python
len(soup.find("tbody").find_all("tr"))
```

```python
# First we isolate the body of the table which contains all the information
# Then we loop through each row and find all the column values for each row
for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text
    #we append the data of each row to table
    netflix_data = pd.concat([netflix_data,pd.DataFrame({'Date':[date],'Open':[open],'High':[high],'Low':[low],'Close':[close],'Adj Close':[adj_close],'Volume':[volume]})])
```

```python
html_data = requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/amazon_data_webpage.html").text

```

```python
data = BeautifulSoup(html_data,"html.parser")
```

```python
data.title.string
```

```python
amazon_df = pd.DataFrame(columns = ['Date','Open','High','Low','Close','Adj Close','Volume'])
for row in data.find('tbody').find_all('tr'):
    col = row.find_all('td')
    open = col[1].text
    high = col[2].text
    low = col[3].text
    close = col[4].text
    adj_close = col[5].text
    volume = col[6].text
    date = col[0].text
    amazon_df = pd.concat([amazon_df,pd.DataFrame({'Date':date,'Open':[open],'High':[high],'Low':[low],'Close':[close],'Adj Close':[adj_close],'Volume':[volume]})])
```
